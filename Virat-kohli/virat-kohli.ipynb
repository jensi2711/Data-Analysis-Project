look at the trend of runs scored by Virat Kohli in his career from 18 August 2008 to 22 January 2017: 

```python
df['Start Date']=pd.to_datetime(df['Start Date'])
df['Start Date']
```

```python
start_date='2008-8-18'
end_date='2017-1-22'
```

```python
trending_runs_score=(df['Start Date']>=start_date) & (df['Start Date']<=end_date)
trending_runs_score
```

```python
filter_data=df.loc[trending_runs_score]
filter_data
```

```python
plt.plot(filter_data['Start Date'], filter_data['Runs'], marker='o', linestyle='-', color='b')
plt.title('Trend of Runs Scored by Virat Kohli (18 Aug 2008 to 22 Jan 2017)',color='r')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.grid()
plt.xticks(rotation=30)
plt.tight_layout()
```

letâ€™s analyze Virat Kohliâ€™s strike rate. 

```python
df['strike']=(df['Runs']/df['BF'])*100
df['strike']
```

```python
df['mean_strike']=df['strike'].mean()
df['mean_strike']
```

```python
opposite_strike=df.groupby('Opposition')['Runs'].mean()
opposite_strike
```

```python
plt.xlabel('opposite')
plt.ylabel('Rate')
plt.title('Strike Rate',color='r')
plt.xticks(rotation=50)
opposite_strike.plot()
```
